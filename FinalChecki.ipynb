```python
import pandas as pd

# Step 1: Load the saved FinalDataset.csv
dataset = pd.read_csv("/home/jibril/Documents/REvoEstatePricePrediction/FinalDataset.csv")

# Step 2: Check each column
print("=== Checking FinalDataset.csv ===")

# Check for missing values in each column
print("\n1. Missing Values (NaN) per Column:")
print(dataset.isnull().sum())

# Check data types and non-null counts
print("\n2. Data Types and Info:")
print(dataset.info())

# Print summary statistics or sample values for each column
print("\n3. Summary for Each Column:")
for column in ["Price", "Bedrooms", "Bathrooms", "Land_Area", "Year", "Status", "Furnished", "Address", "Property_Type"]:
    print(f"\nColumn: {column}")
    if dataset[column].dtype in ['float64', 'int64']:  # Numerical columns
        print(dataset[column].describe())  # Stats like mean, min, max
    else:  # Categorical columns
        print(dataset[column].value_counts())  # Frequency of each value

# Step 3: Print the full loaded dataset
print("\n4. Loaded FinalDataset.csv:")
print(dataset)
```
